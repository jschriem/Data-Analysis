# Sun Run Training Data
From April 22, 2018 to April 14, 2019 I recorded most of my running data, and thought it would be cool to continue practicing pandas and matplotlib by using some relevant data.

The running logs consist of a date, total duration, distance, and the pace.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

run = pd.read_csv('./sunrunCSV.csv')

print(run.head())
```

```python
# Distance Run
fig,ax = plt.subplots()

distance = plt.bar(run['Date'], run['Distance(KM)'])

plt.title('Distance Ran Between 10km Races')
plt.xlabel('Date (m/d/y)')
plt.ylabel('Distance (KM)')
plt.xticks(rotation=45)


distance_mean = [np.mean(run['Distance(KM)']) for i in run['Distance(KM)']]

mean_line = ax.plot(run['Date'] ,distance_mean, label='Mean', linestyle='--')
```

```python
# Pace Run
fig,ax = plt.subplots()

pace = plt.plot(run['Date'], run['Pace(Time/KM)'],linestyle='-', marker='o')

plt.xticks(rotation=45)
plt.title('Pace Between 10KM Races')
plt.xlabel('Date (m/d/y)')
plt.ylabel('Pace (Time per KM)')

pace_mean = [np.mean(run['Pace(Time/KM)']) for i in run['Pace(Time/KM)']]
mean_line2 = ax.plot(run['Date'] ,pace_mean, label='Mean', linestyle='--')
```

```python
# Distance vs Pace
fig,ax = plt.subplots()
plt.xlabel('Date (m/d/y)')
t = run['Date']
plt.xticks(rotation=45)
ax.set_title('Distance and Pace')

#First plot
ax.set_ylabel('Distance (KM)', color='r')
plot1 = ax.plot(t, run['Distance(KM)'], marker='.', color='r')

#Second plot
ax2 = ax.twinx()
ax2.set_ylabel('Pace(Time/KM)', color='b')
plot2 = ax2.plot(t, run['Pace(Time/KM)'], marker='.', color='b')


#Vertical lines to seperate months
plt.axvline(x=1,color = '#CFE2EE', label = 'Month')
plt.axvline(x=3,color = '#CFE2EE')
plt.axvline(x=5,color = '#CFE2EE')
plt.axvline(x=7,color = '#CFE2EE')
plt.axvline(x=8,color = '#CFE2EE')
plt.axvline(x=13,color = '#CFE2EE')
plt.axvline(x=16,color = '#CFE2EE')
plt.axvline(x=19,color = '#CFE2EE')


#legend
plt.legend(loc='upper center')

#show plots
plt.show() 
```
